```python
from selenium import webdriver
import time
import pandas as pd
```

```python
#자신의 chromedriver가 있는 파일 위치를 path 변수에 할당
#window는 ~/chromedriver.exe를 불러오고 mac은 exe를 빼고 불러오기

path = r"C:\Users\User\Downloads\chromedriver_win32\chromedriver.exe"
```

```python
#짠! selenium으로 제어할 수 있는 브라우저 새 창이 뜬다.
driver = webdriver.Chrome(path)
```

```python
# 더본코리아 매장정보 페이지 들어가기
driver.get('http://www.theborn.co.kr/store/domestic-store/')
```

```python
storelists = driver.find_elements_by_css_selector('.map_body > div')

for div in storelists:
    temp = div.find_element_by_css_selector('ul.map_more2').text.split()
    print('브랜드 : ', temp[0])
    print('지역 : ', temp[1])
    print('매장명 : ', temp[2])
    print('주소 : ', ' '.join(temp[3:-2]))
    print('*'*100)
    
driver.find_element_by_xpath('//*[@id="pagination"]/li[6]').click()
time.sleep(3)
storelists = driver.find_elements_by_css_selector('.map_body > div')
time.sleep(3)
storelists = driver.find_elements_by_css_selector('.map_body > div')

for div in storelists:
    temp = div.find_element_by_css_selector('ul.map_more2').text.split()
    print('브랜드 : ', temp[0])
    print('지역 : ', temp[1])
    print('매장명 : ', temp[2])
    print('주소 : ', ' '.join(temp[3:-2]))
    print('*'*100)

driver.find_element_by_xpath('//*[@id="pagination"]/li[7]').click()
time.sleep(3)
storelists = driver.find_elements_by_css_selector('.map_body > div')
time.sleep(3)
storelists = driver.find_elements_by_css_selector('.map_body > div')


for i in range(142):
    for div in storelists:
        temp = div.find_element_by_css_selector('ul.map_more2').text.split()
        print('브랜드 : ', temp[0])
        print('지역 : ', temp[1])
        print('매장명 : ', temp[2])
        print('주소 : ', ' '.join(temp[3:-2]))
        print('*'*100)
    time.sleep(3)
    driver.find_element_by_xpath('//*[@id="pagination"]/li[8]').click()
    time.sleep(3)
    storelists = driver.find_elements_by_css_selector('.map_body > div')
    time.sleep(3)
    storelists = driver.find_elements_by_css_selector('.map_body > div')
    time.sleep(3)

time.sleep(3)

for div in storelists:
    temp = div.find_element_by_css_selector('ul.map_more2').text.split()
    print('브랜드 : ', temp[0])
    print('지역 : ', temp[1])
    print('매장명 : ', temp[2])
    print('주소 : ', ' '.join(temp[3:-2]))
    print('*'*100)
```

```python
#list와 dictionary 초기화
brand = []
region = []
store_name = []
addr = []
colname = ['브랜드', '지역', '매장명', '주소']
```

```python
# 매장 하나하나 접근
for div in storelists:
    temp = div.find_element_by_css_selector('ul.map_more2').text.split()
    brand.append(temp[0])
    region.append(temp[1])
    store_name.append(temp[2])
    addr.append((' '.join(temp[3:-2])))
```

```python
store_list_DF = pd.DataFrame([brand,region,store_name,addr]).T
store_list_DF.columns = colname
store_list_DF
```

```python
store_list_DF.to_csv('1차 크롤링.csv', index=False)
```
